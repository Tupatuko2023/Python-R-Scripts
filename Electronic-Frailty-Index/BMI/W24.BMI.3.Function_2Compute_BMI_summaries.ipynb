```python
# Work 24: Function to Compute and Print Cohort Summaries of Calculated BMI Data:
#  [W24.BMI.3.Function_2Compute_BMI_summaries.ipynb]

# "This Jupyter notebook script loads BMI data, defines a function to compute cohort summaries, and prints
#  the calculated summaries for further analysis."

########################################################################################################
#  Sequence list
########################################################################################################

# 1: Load the calculated data
# 2: Print column names for debugging
# 3: Function to compute cohort summaries
# 4: Compute and print cohort summaries
# 5: Further analysis can be added here if needed

########################################################################################################
########################################################################################################

import pandas as pd

# 1: Load the calculated data
calculated_path = "/home/work/BMI_calculated.csv"
calculated_data = pd.read_csv(calculated_path, sep="|")
print("1: Loaded the calculated data.")

# 2: Print column names for debugging
print("2: Column names in the calculated data:")
print(calculated_data.columns)


# 3: Function to compute cohort summaries
def compute_cohort_summaries(data, column="BMI"):
    summary = {
        "min": data[column].min(),
        "max": data[column].max(),
        "mean": data[column].mean(),
        "std": data[column].std(),
        "count": data[column].count(),
    }
    return summary


print("3: Created a function to compute cohort summaries.")

# 4: Compute and print cohort summaries
bmi_summary = compute_cohort_summaries(calculated_data)
print("4: Cohort summaries for BMI records:")
print(f"Min BMI: {bmi_summary['min']:.2f}")
print(f"Max BMI: {bmi_summary['max']:.2f}")
print(f"Mean BMI: {bmi_summary['mean']:.2f}")
print(f"Standard Deviation BMI: {bmi_summary['std']:.2f}")
print(f"Count of BMI records: {bmi_summary['count']}")

# 5: Further analysis can be added here if needed
# For example, you can segment the data based on different cohorts such as age, gender, etc.,
# and calculate summaries for each segment.
```
